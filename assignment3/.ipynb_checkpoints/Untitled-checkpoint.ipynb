```python
import pandas as pd
import matplotlib.pyplot as plt
import pylab
import matplotlib
import numpy as np
```

```python
history = pd.read_csv('web_count.csv')
history = history.sort_values(by='count', ascending=False)
history_filtered = history[9:29]
history_filtered
```

```python
history_filtered.plot(kind='barh', title='Deaths for under 5yrs of age for developed region in 2000 and 2015', rot=1, color=(['m']),
                  figsize=(11, 9), edgecolor='w', linewidth=2, width=0.7)
plt.xlabel('<------ Total count ------>', fontsize=14, color='m')
plt.show()
```
